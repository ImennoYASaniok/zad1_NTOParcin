```python
import pandas as pd
import numpy as np

dt_test = pd.read_csv('raw_data_0a14f818-b0de-11ec-b89c-e82aea2c97f4.csv')

test_col = dt_test["WRIST.x"]
new_col = []
dop_new_col = []

print(f"ДО ({test_col.shape[0]}):\n", test_col)

ind_delit = 5
rasnos = 300
old_el = test_col[0]
ind_test_col = 0
for el in test_col:
    #print(float(str(round(el, ind_delit))[:-1]), float(str(round(old_el, ind_delit))[:-1]))
    if ind_test_col > 0 and round(abs(float(str(el)[:ind_delit+2]) - float(str(old_el)[:ind_delit+2])), ind_delit)*pow(10, ind_delit+1) < rasnos:
        test_per = abs(el - old_el) * pow(10, ind_delit+1)
        print(old_el, el, test_per, rasnos)
        dop_new_col.append(el)
    elif ind_test_col > 0:
        #print(np.array(dop_new_col))
        new_col.append(np.mean(np.array(dop_new_col)))
        dop_new_col.clear()
    old_el = el
    ind_test_col += 1

print(f"После ({len(new_col)}):\n", new_col)


```
